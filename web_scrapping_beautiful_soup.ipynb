```python
#### Import required libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
##### url of websites
url = 'https://www.ambitionbox.com/list-of-companies?page=1'
```

```python
#### Access the websites through HTTPrequests
webpage = requests.get(url).text
```

```python
### Create a variable and call the Beautifulsoup function
soup = BeautifulSoup(webpage, 'lxml' )
```

```python
##### Arrange the file in right format
print(soup.prettify())
```

```python
##### To find the content of website i.e h1
soup.find_all('h1')[0].text
```

```python
##### To scrap the title of the company
soup.find_all('h2')
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
#### To scrap rating of the company
for i in soup.find_all('p',class_='rating'):
    print(i.text.strip())
    
```

```python
##### To scrap the review of the company
soup.find_all('a',class_='review-count')
```

```python
for i in soup.find_all('a',class_='review-count'):
    print(i.text.strip())
```

```python
#### To scrap the type of company i.e either public or private
soup.find_all('p',class_='infoEntity')
```

```python
for i in soup.find_all('p',class_='infoEntity'):
    print(i.text.strip())
```

```python
###### To scrap the whole container of company
container = soup.find_all('div',class_='company-content-wrapper')
```

```python
len(container)
```

```python
#### Scrapping the content of the company through container by using for loop
title = []
rating = []
review = []
type = []
headquarter = []
years = []
employees = []
description = []
for i in container:
    title.append(i.find('h2').text.strip())     ###### to scrap the name of the company
    rating.append(i.find('p',class_='rating').text.strip()) ##### to scrap the rating of the company
    review.append(i.find('a',class_='review-count').text.strip()) ##### to scrap the review of the company
    type.append(i.find_all('p',class_='infoEntity')[0].text.strip())  #### to scrap the type of the company
    headquarter.append(i.find_all('p', class_='infoEntity')[1].text.strip()) #### To scrap the headquarter of the company
    years.append(i.find_all('p', class_ = 'infoEntity')[2].text.strip()) ###### To scrap the years of the company
    employees.append(i.find_all('p', class_='infoEntity')[3].text.strip()) ##### To scrap the No.of Employees of the company
    description.append(i.find('p', class_='description').text.strip()) #### To scrap the description of the company
```

```python
##### To creating the dataset
df = {
    'title':title,
    'rating':rating,
    'review':review,
    'type':type,
    'headqaurter':headquarter,
    'years':years,
    'employees':employees,
    'description':description
}
```

```python
df = pd.DataFrame(df)
```
